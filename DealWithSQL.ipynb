```python
import pandas as pd
```

```python
import sqlite3
```

## Read data from SQL database with Python

```python
# step 1: connect to the database

conn = sqlite3.connect('data/chinook.db')

# step 2: create a cursor that allows us to execute SQL queries against a database

cur = conn.cursor()

# step 3: use the method execute to receive SQL parameters to run against the database

cur.execute('SELECT * FROM employees LIMIT 5;') # fetch the first 5 rows from the employees table

# step 4: use fetchall method to fetch all of the results of a query


results = cur.fetchall()


results
```

```python
# manually add column headers
df = pd.DataFrame(results)

df.head()
```

## Use Pandas read_sql method

```python
conn = sqlite3.connect('data/chinook.db')

df = pd.read_sql('SELECT * FROM employees;', conn)

df.head()
```

```python

df = pd.read_sql('SELECT * FROM employees;', conn,
                 index_col='EmployeeId',
                 parse_dates=['BirthDate', 'HireDate'])


df.head()
```
