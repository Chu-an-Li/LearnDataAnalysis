## Handling Missing Data with Pandas

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Column A': [1, np.nan, 30, np.nan],
    'Column B': [2, 8, 31, np.nan],
    'Column C': [np.nan, 9, 32, 100],
    'Column D': [5, 8, 34, 110],
})
```

```python
df
```

```python
df.isnull()
```

```python
df.notnull()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.dropna(axis=1)  # axis='columns' also works
```

```python
df.dropna(thresh=3) # only drop it if there are 3 NaN
```

```python
# Filling null values

# with an arbitrary value
s = ...
s.fillna(0)
s.fillna(s.mean())
```

```python
# Filling nulls with contiguous (close) values

s.fillna(method='ffill') # from front
s.fillna(method='bfill') # from behind
```

```python
df.fillna(method='ffill', axis=0)
```
